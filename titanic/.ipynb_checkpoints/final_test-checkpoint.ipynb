```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv("train.csv")
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.head()
```

```python
df['Cabin_2'].value_counts()
```

```python
c = df.groupby('Cabin_2').sum()['Survived'] / df['Cabin_2'].value_counts()
```

```python
df['Cabin_3'] = df["Cabin_2"].map(c.to_dict())
```

```python
df.set_index("PassengerId",inplace=True)
```

```python
df.head()
```

```python
g = sns.FacetGrid(df,col = "Survived")
g.map(plt.hist,"Age",bins=20)
```

```python
def func(x):
    if x['Age']<8:return "child"
    if x['Age']>55:return "old"
    else :return x['Sex']

df['Sex_2'] = df.apply(func,axis=1)
```

```python
df.groupby("Sex_2")['Survived'].sum() / df.Sex_2.value_counts()
```

```python
df.head()
```

```python
sns.factorplot(x="Embarked",y="Survived",data=df,kind='bar')
```

```python
df.Embarked.value_counts()
df['Embarked_2'] = df['Embarked'].map({"C":1,"Q":2,"S":3})
```

```python
df.head()
```

```python
values = ['Pclass',"Age","Parch","Fare","Cabin_3","Sex_3","Embarked_2"]
train = df[values]
test = df.Survived
train.Age.fillna(30,inplace=True)
train.Embarked_2.fillna(2,inplace=True)
```

```python
train.head()
```

```python
train.Embarked_2.fillna('Q',inplace=True)
train['Embarked_2']
```

```python
clf = LogisticRegression()
clf.fit(train[0:-50],test[0:-50])
```

```python
clf.predict(train[-50:])
```

```python
clf.score(train[-50:],test[-50:])
```
